```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
page=requests.get('https://www.imdb.com/search/title/?title_type=feature&year=2022-01-01,2022-12-31&sort=num_votes,desc')
soup=BeautifulSoup(page.content,'lxml')
```

```python
Title=soup.findAll('h3',class_='lister-item-header')
MovieTitle=[]
for i in Title:
    j=i.a.text
    MovieTitle.append(j)
print(MovieTitle)
```

```python
r=soup.findAll('div',class_='inline-block ratings-imdb-rating')
Rating=[]
for i in r:
    j=i.strong.text
    Rating.append(j)
print(Rating)
```

```python
genre=soup.findAll('span',class_='genre')
Genre=[]
for i in genre:
    j=i.text.strip()
    Genre.append(j)
print(*Genre)
```

```python
certification=soup.find_all('span',class_='certificate')
Certification=[]
for i in certification:
    j=i.text
    Certification.append(j)
print(Certification)
```

```python
runtime=soup.findAll('span',class_='runtime') 
RunTime=[]
for i in runtime:
    j=i.text
    RunTime.append(j)
print(RunTime)
```

```python
votes=soup.find_all('p',class_='sort-num_votes-visible')
Votes=[]
for i in votes:
    j=i.text[7:].strip()
    Votes.append(j)
print(Votes)
```

```python
## saving data in Csv ##
import numpy as np
import pandas as pd
data={'MovieTitle':MovieTitle,'Rating':Rating,'Genre':Genre,'Certification':Certification,'RunTime':RunTime,'Votes':Votes}
df=pd.DataFrame(data=data)
df
```
